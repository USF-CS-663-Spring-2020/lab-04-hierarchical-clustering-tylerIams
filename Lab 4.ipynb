```python
import pandas as pd
```

Let's read in the data and check it out.  Looks like it has nearly 2000 observations and 69 features

```python
df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df, 'ward')
```

Now we've got Z, so let's draw the dendrogram for the first dataset

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(Z)
plt.show()
```

Now let's read in the second dataset, this one is a lot bigger, with over 11,000 observations and over 5,800 features

```python
df2 = pd.read_csv('NIPS_1987-2015.csv')
```

```python
df2 = df2.drop('Unnamed: 0', 1)
```

```python
Z2 = linkage(df2, 'ward')
```

```python
%matplotlib inline

plt.figure()
dendrogram(Z2)
plt.show()
```
